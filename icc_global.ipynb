Combined figure

```python
import plotly.graph_objects as go
import plotly.subplots as sp
import plotly.express as px  # for colors
import re

def color_to_rgba(color, alpha=1.0):
    # If already rgba, just return
    if color.startswith('rgba'):
        return color
    # If already rgb, add alpha
    if color.startswith('rgb'):
        nums = re.findall(r'\d+', color)
        return f'rgba({nums[0]},{nums[1]},{nums[2]},{alpha})'
    # If hex, convert
    color = color.lstrip('#')
    return f'rgba({int(color[0:2],16)},{int(color[2:4],16)},{int(color[4:6],16)},{alpha})'


# Your ICC values and confidence intervals 
icc_values = [
    r1_marias_icc, r2_marias_icc, val_marias_icc, 
    r1_mariae_icc, r2_mariae_icc, val_mariae_icc, 
    r1_lemann_icc, r2_lemann_icc, val_lemann_icc
]  


lower_bounds = [
    r1_marias_lower, r2_marias_lower, val_marias_lower, 
    r1_mariae_lower, r2_mariae_lower, val_mariae_lower, 
    r1_lemann_lower, r2_lemann_lower, val_lemann_lower
]  

upper_bounds = [
    r1_marias_upper, r2_marias_upper, val_marias_upper, 
    r1_mariae_upper, r2_mariae_upper, val_mariae_upper, 
    r1_lemann_upper, r2_lemann_upper, val_lemann_upper
]  

# Calculate asymmetric errors
errors_lower = [icc_values[i] - lower_bounds[i] for i in range(len(icc_values))]
errors_upper = [upper_bounds[i] - icc_values[i] for i in range(len(icc_values))]

# Grouping labels and colors
x_labels = ['Round 1', 'Round 2', 'Validation set']
groups = ['MaRIAs', 'MaRIAs-E', 'Lemann']
colors = px.colors.qualitative.D3[:3]  # Pick 3 distinct colors

# Create subplot layout: 1 row, 3 cols
fig = sp.make_subplots(rows=1, cols=3, subplot_titles=groups)

for i, group in enumerate(groups):
    y_vals = icc_values[i*3:(i+1)*3]
    err_lower = errors_lower[i*3:(i+1)*3]
    err_upper = errors_upper[i*3:(i+1)*3]
    ci_lower = lower_bounds[i*3:(i+1)*3]
    ci_upper = upper_bounds[i*3:(i+1)*3]

    hover_text = [
        f"ICC = {y:.2f}<br>95% CI = [{ci_lower:.2f}, {ci_upper:.2f}]"
        for y, ci_lower, ci_upper in zip(y_vals, ci_lower, ci_upper)
    ]

    fig.add_trace(go.Bar(
        x=x_labels,
        y=y_vals,
        name=group,
        error_y=dict(
            type='data',
            symmetric=False,
            array=err_upper,
            arrayminus=err_lower,
            thickness=2,
            width=5
        ),
        marker=dict(
            color=color_to_rgba(colors[i], 0.55),
            line=dict(color=colors[i], width=2)
        ),
        hovertext=hover_text,
        hoverinfo='text',
        # Remove text & textposition to avoid inconsistent font sizing
    ), row=1, col=i+1)

    # Add annotations for ICC value and CI bounds
    for j, label in enumerate(x_labels):
        icc_y = y_vals[j]
        upper_y = ci_upper[j]
        lower_y = ci_lower[j]

        # ICC value (on top of bar)
        fig.add_annotation(
            x=label,
            y=icc_y,
            text=f"{icc_y:.2f}",
            showarrow=False,
            yshift=10,  # small bump above bar
            font=dict(size=11),
            row=1,
            col=i+1
        )

        # Upper CI bound
        fig.add_annotation(
            x=label,
            y=upper_y,
            text=f"{upper_y:.2f}",
            showarrow=False,
            yshift=8,
            font=dict(size=11),
            row=1,
            col=i+1
        )

        # Lower CI bound
        fig.add_annotation(
            x=label,
            y=lower_y,
            text=f"{lower_y:.2f}",
            showarrow=False,
            yshift=-10,
            font=dict(size=11),
            row=1,
            col=i+1
        )


    # Set y-axis title only on first subplot
    if i == 0:
        fig.update_yaxes(title_text="ICC", row=1, col=1)
    fig.update_xaxes(row=1, col=i+1)

# Add horizontal dotted line at y=0.5 in all subplots
for i in range(1, 4):
    fig.add_shape(
        type="line",
        x0=-0.5, x1=2.5,
        y0=0.5, y1=0.5,
        xref=f"x{i}",
        yref=f"y{i}",
        line=dict(color="gray", width=2, dash="dot"),
    )

fig.update_layout(
    template="simple_white",
    width=1000, height=500,
    showlegend=False,
    uniformtext_minsize=8,
    uniformtext_mode='hide'
)
fig.update_yaxes(range=[0, 1], showgrid=True)  
fig.update_yaxes(tickvals=np.arange(0, 1.1, 0.1), ticktext=[f"{i:.1f}" for i in np.arange(0, 1.1, 0.1)])

fig.show()
```

#### Score distribution

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.subplots as sp

# --- Load and process data (unchanged) ---
maria1 = pd.read_csv("Inputs/round_1_maria_global.csv", header=0)
lemann1 = pd.read_csv("Inputs/round_1_lemann_global.csv", header=0)

maria2 = pd.read_csv("Inputs/round_2_maria_global.csv", header=0)
lemann2 = pd.read_csv("Inputs/round_2_lemann_global.csv", header=0)

maria3 = pd.read_csv("Inputs/validation_maria_global.csv", header=0)
maria3['patientIdentifier'] = maria3['patientIdentifier'].astype(str)
maria3['procedureName'] = maria3['procedureName'].astype(str)
maria3['scan'] = maria3['patientIdentifier'] + ' ' + maria3['procedureName']
maria3 = maria3[maria3['scan'].map(maria3['scan'].value_counts()) > 1]

lemann3 = pd.read_csv("Inputs/validation_lemann_global.csv", header=0)
lemann3['patientIdentifier'] = lemann3['patientIdentifier'].astype(str)
lemann3['procedureName'] = lemann3['procedureName'].astype(str)
lemann3['scan'] = lemann3['patientIdentifier'] + ' ' + lemann3['procedureName']
lemann3 = lemann3[lemann3['scan'].map(lemann3['scan'].value_counts()) > 1]

# Annotate rounds
maria1['Round'] = maria2['Round'] = maria3['Round'] = lemann1['Round'] = lemann2['Round'] = lemann3['Round'] = None
maria1['Round'] = lemann1['Round'] = 'Round 1'
maria2['Round'] = lemann2['Round'] = 'Round 2'
maria3['Round'] = lemann3['Round'] = 'Validation'

# Combine
marias_combined = pd.concat([maria1[['mariaS', 'Round']], maria2[['mariaS', 'Round']], maria3[['mariaS', 'Round']]])
marias_combined.rename(columns={'mariaS': 'Score'}, inplace=True)
marias_combined['Metric'] = 'MaRIAs'

mariae_combined = pd.concat([maria1[['mariaE', 'Round']], maria2[['mariaE', 'Round']], maria3[['mariaE', 'Round']]])
mariae_combined.rename(columns={'mariaE': 'Score'}, inplace=True)
mariae_combined['Metric'] = 'MaRIAs-E'

lemann_combined = pd.concat([lemann1[['lemannIndex', 'Round']], lemann2[['lemannIndex', 'Round']], lemann3[['lemannIndex', 'Round']]])
lemann_combined.rename(columns={'lemannIndex': 'Score'}, inplace=True)
lemann_combined['Metric'] = 'Lemann'

# Stack all
df_all = pd.concat([marias_combined, mariae_combined, lemann_combined])

# Create 3 rows  3 columns subplots
fig = sp.make_subplots(
    rows=3, cols=3,
    subplot_titles=[
        f"MaRIAs - {r}" for r in ['Round 1', 'Round 2', 'Validation']
    ] + [
        f"MaRIAs-E - {r}" for r in ['Round 1', 'Round 2', 'Validation']
    ] + [
        f"Lemann - {r}" for r in ['Round 1', 'Round 2', 'Validation']
    ]
)

# Add a histogram for each Metric-Round combo
metrics = ['MaRIAs', 'MaRIAs-E', 'Lemann']
rounds = ['Round 1', 'Round 2', 'Validation']

for i, metric in enumerate(metrics):
    for j, round_ in enumerate(rounds):
        subset = df_all[(df_all['Metric'] == metric) & (df_all['Round'] == round_)]
        fig.add_trace(go.Histogram(
            x=subset['Score'],
            nbinsx=20,
            marker_color='black',
            showlegend=False
        ), row=i+1, col=j+1)

# Formatting
fig.update_layout(
    template="simple_white",
    height=900,
    width=1000,
    title=""
)

# Consistent axes
for r in range(1, 4):
    for c in range(1, 4):
        fig.update_xaxes(range=[0, 11], row=r, col=c)
        fig.update_yaxes(showgrid=True, row=r, col=c)

fig.show()
```

#### Kruskal-Wallis

```python

# Kruskal-Wallis 
kw_results = pd.DataFrame(columns=['MaRIAs', 'MaRIAs-E', 'Lemann'], index=['H statistic', 'p-value'])

# check for normality
# print('Normality tests:')
norm = pg.normality(data=marias_combined, dv='MaRIAs', group='Round')
# print('MaRIAs', norm)
norm = pg.normality(data=mariae_combined, dv='MaRIAs-E', group='Round')
# print('MaRIAs-E', norm)
norm = pg.normality(data=lemann_combined, dv='Lemann', group='Round')
# print('Lemann', norm)

# check for homogeneity of variance
# print('\nHomogeneity of variance tests:')
hom = pg.homoscedasticity(data=marias_combined, dv='MaRIAs', group='Round', method='levene')
# print('MaRIas', hom)
hom = pg.homoscedasticity(data=mariae_combined, dv='MaRIAs-E', group='Round', method='levene')
# print('MaRIas-E', hom)
hom = pg.homoscedasticity(data=lemann_combined, dv='Lemann', group='Round', method='levene')
# print('Lemann', hom)

# perform kruskal wallis test to check for differences between the three rounds
kw = pg.kruskal(data=marias_combined, dv='MaRIAs', between='Round')
print('MaRIAs', kw)
kw_results.loc['H statistic', 'MaRIAs'] = kw['H'][0]
kw_results.loc['p-value', 'MaRIAs'] = kw['p-unc'][0]

kw = pg.kruskal(data=mariae_combined, dv='MaRIAs-E', between='Round')
print('MaRIAs-E', kw)
kw_results.loc['H statistic', 'MaRIAs-E'] = kw['H'][0]
kw_results.loc['p-value', 'MaRIAs-E'] = kw['p-unc'][0]

kw = pg.kruskal(data=lemann_combined, dv='Lemann', between='Round')
print('Lemann', kw)
kw_results.loc['H statistic', 'Lemann'] = kw['H'][0]
kw_results.loc['p-value', 'Lemann'] = kw['p-unc'][0]
```
